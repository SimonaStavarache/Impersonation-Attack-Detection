```python
c = list()
for i in range(len(data.columns)):
    if(len(np.unique(data.iloc[:,i])) == 1):
        c.append(data.columns[i])
        
data.drop(c,axis = 1, inplace = True)
```

```python
data.describe()
```

```python
correlations = data.corr()
# plot correlation matrix after removal of unique values
fig = plt.figure()
ax = fig.add_subplot(111)
cax = ax.matshow(correlations, vmin=-1, vmax=1)
fig.colorbar(cax)
plt.show()
```

```python
pearsoncorr = data.corr(method='pearson')
pearsoncorr
```

```python
sb.heatmap(pearsoncorr, 
            xticklabels=pearsoncorr.columns,
            yticklabels=pearsoncorr.columns,
            cmap='RdBu_r',
            annot=True)
```

```python
spearmancorr = data.corr(method='spearman')
spearmancorr
```

```python
from scipy import stats
plt.figure(figsize=(9,7))
sb.distplot(data['frame.time_delta_displayed'],fit=stats.norm)
plt.legend(['skew{:.2f}'.format(data['class'].skew())],loc='best')
plt.ylabel('Frequency')
plt.title('Target distribution');
```
