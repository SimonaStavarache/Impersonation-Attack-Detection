```python
X = data.iloc[:,1:]
y = data.iloc[:,0]
feature_names = data.drop('class', 1).columns
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Feature selection using LASSO
from sklearn.linear_model import Lasso
from sklearn.model_selection import GridSearchCV

grid_values = {'alpha': [0.0235, 0.024, 0.0245]}
grid_lasso = GridSearchCV(Lasso(), param_grid = grid_values, cv=10, scoring = 'r2')
grid_lasso.fit(X_train, y_train)
predict_lasso = grid_lasso.predict(X_test)

print('Mean score matrix: ', grid_lasso.cv_results_['mean_test_score'])
print('Grid best parameter (max. accuracy): ', grid_lasso.best_params_)
print('Grid best score (accuracy): ', grid_lasso.best_score_)
```

```python
from sklearn import ensemble
model = ensemble.RandomForestRegressor(n_estimators=200, max_depth=10, min_samples_leaf=4, max_features=0.2, n_jobs=-1, random_state=0)
model.fit(X_train,y_train)
feat_names = X_train.columns.values

## plot the importances ##
importances = model.feature_importances_
std = np.std([tree.feature_importances_ for tree in model.estimators_], axis=0)
indices = np.argsort(importances)[::-1][:20]

plt.figure(figsize=(12,12))
plt.title("Feature importances")
plt.bar(range(len(indices)), importances[indices], color="r", align="center")
plt.xticks(range(len(indices)), feat_names[indices], rotation='vertical')
plt.xlim([-1, len(indices)])
plt.show();

print(model.score(X_test,y_test))
```

```python
# applying RandomForest using only the most important features
pls = data.drop(feat_names, 1).columns
X_train_subset = X_train[feat_names[indices]]
# print(X_train_subset)
new_model = ensemble.RandomForestRegressor(n_estimators=200, max_depth=10, min_samples_leaf=4, max_features=0.2, n_jobs=-1, random_state=0, warm_start = True)
new_model.fit(X_train_subset, y_train)
X_test_part = X_test[feat_names[indices]]
new_model_predict = new_model.predict(X_test_part)

print("Accuracy: " + str(new_model.score(X_test_part, y_test)))
```

```python
# applying DecisionTreeClassifier on the most important features
from sklearn.tree import DecisionTreeRegressor
model_DecTree = DecisionTreeRegressor()

model_DecTree.fit(X_train, y_train)
model_DecTree_predict = model_DecTree.predict(X_test)
print("Accuracy without feature selection: " + str(model_DecTree.score(X_test, y_test)))

model_DecTree.fit(X_train_subset, y_train)
model_DecTree_predict = model_DecTree.predict(X_test_part)

print("Accuracy with feature selection: " + str(model_DecTree.score(X_test_part, y_test)))
```
