```python
# chaged this two variables: X must be all the other features and y the class column (the last column)
X = data.iloc[:,0:-1]
y = data.iloc[:,-1]

feature_names = data.drop('class', 1).columns
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Feature selection using LASSO
from sklearn.linear_model import Lasso
from sklearn.model_selection import GridSearchCV

grid_values = {'alpha': [0.0235, 0.024, 0.0245]}
grid_lasso = GridSearchCV(Lasso(), param_grid = grid_values, cv=10, scoring = 'r2')
grid_lasso.fit(X_train, y_train)
predict_lasso = grid_lasso.predict(X_test)

print('Mean score matrix: ', grid_lasso.cv_results_['mean_test_score'])
print('Grid best parameter (max. accuracy): ', grid_lasso.best_params_)
print('Grid best score (accuracy): ', grid_lasso.best_score_)
```

```python
from sklearn import ensemble
model = ensemble.RandomForestRegressor(n_estimators=200, max_depth=10, min_samples_leaf=4, max_features=0.2, n_jobs=-1, random_state=0)
model.fit(X_train, y_train)
feat_names = X_train.columns.values

## plot the importances ##
importances = model.feature_importances_
std = np.std([tree.feature_importances_ for tree in model.estimators_], axis=0)
indices = np.argsort(importances)[::-1][:20]

plt.figure(figsize=(12,12))
plt.title("Feature importances")
plt.bar(range(len(indices)), importances[indices], color="r", align="center")
plt.xticks(range(len(indices)), feat_names[indices], rotation='vertical')
plt.xlim([-1, len(indices)])
plt.show();

```

```python
def preprocessing (file):
    # reading the data
    data = pd.read_csv(file,names = col_names)
    
    # Checking and droping Unique values
    c = list()
    for i in range(len(data.columns)):
        if(len(np.unique(data.iloc[:,i])) == 1):
            c.append(data.columns[i])
        
    data.drop(c,axis = 1, inplace = True)

    X = data.iloc[1:,0:-1]
    y = data.iloc[1:,-1]
    return X, y
    
```

```python
data_test = pd.read_csv('test.csv',names = col_names)

test_X = data_test.iloc[1:,0:-1]
test_X_part = test_X[feat_names[indices]]
test_y = data_test.iloc[1:,-1]

# test_X, test_y = preprocessing('test.csv')
# test_X_part = test_X[feat_names[indices]]
import time
```
